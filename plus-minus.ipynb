```python
# Implementing plus and minus states and measuring them

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram

kit_0 = Statevector([1, 0])
circuit_1 = QuantumCircuit(1)

circuit_1.h(0)

plus_v = kit_0.evolve(circuit_1)
statistics = plus_v.sample_counts(8000)
plot_histogram(statistics)
```

```python
kit_1 = Statevector([0, 1])
circuit_2 = QuantumCircuit(1)

circuit_2.h(0)

minus_v = kit_1.evolve(circuit_2)
statistics = minus_v.sample_counts(8000)
plot_histogram(statistics)
```
